```python
from pyspark.sql import SparkSession as SS
spark = SS.builder.appName('HMV').getOrCreate()
```

```python
spark
```

```python
df_spark = spark.read.csv('city_day.csv',header=True,inferSchema=True)
```

```python
df_spark.show()
```

```python
# Dropping Cloumn by name
df_spark = df_spark.drop('AQI_Bucket')
```

```python
df_spark.show()
```

```python
# Dropping Rows with NULL values using how={'all','any'} DEFAULT-'any'
df_spark.na.drop(how='any').show()
df_spark.na.drop().count()
```

```python
df_spark.na.drop(how='all').show()
df_spark.na.drop(how='all').count()
```

```python
df_spark.count()
```

```python
# Dropping Rows with Threshold ie. minimum no. of non null values
df_spark.na.drop(how='any',thresh = 5).show()
print(df_spark.na.drop(how='any',thresh = 5).count())
print(len(df_spark.columns))
```

```python
# Dropping Rows with Subset='column name' ie. from a attribute like axis=columns in pandas
df_spark.na.drop(how='any',subset='PM25').show()
print(df_spark.na.drop(how='any',subset=['PM25']).count())
print(len(df_spark.columns))
```

```python
# Filling the missing Values fill('value',['cl1','cl2',...]) wrt dtype
df_spark.na.fill(9999).show()
```

```python
df_spark.show()
```

```python
# using Imputer from PySpark.ML to transform nulls in each attribute to their respective mean 
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['PM25','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene'],
    outputCols=[f'{c}_imputed' for c in ['PM25','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene'] ]
).setStrategy('mean')

imputer.fit(df_spark).transform(df_spark).show()
```

```python
df_spark.describe().show()
```

```python
imputer = Imputer(
    inputCols=['PM25','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI'],
    outputCols=['PM25','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI'] 
).setStrategy('mean')

imputer.fit(df_spark).transform(df_spark).show()
```
